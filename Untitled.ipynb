```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
path = "C:/Users/sgove/OneDrive/Documents/R_ProgrammingDocs/25_onwards.xlsx"
df = pd.read_excel(path,header=0)
```

```python
df.columns
```

```python
df.tail(3)
```

```python
df1 = df[["Date","Deaths"]]
df1.head()
```

```python
# plot the time series
df1['Deaths'].plot()
plt.show()
```

```python
# prepare expected column names
df1.columns = ['ds', 'y']
df1['ds']= pd.to_datetime(df1['ds'])
```

```python
from fbprophet import Prophet
# define the model
model = Prophet()
# fit the model
model.fit(df1)
```

```python
prediction_size = 90
# define the period for which we want a prediction
future = model.make_future_dataframe(prediction_size)
future.tail()
```

```python
df = df.reset_index(drop=True).dropna()
df1 = df1.reset_index(drop=True).dropna()
future = future.reset_index(drop=True).dropna()
```

```python
forecast = model.predict(future)
forecast.tail()
```

```python
model.plot_components(forecast)
```

```python
model.plot(forecast,xlabel = 'Date', ylabel = 'Deaths')
```

```python
import plotly.graph_objs as go
import plotly.offline as py
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
trace = go.Scatter(
    name = 'Actual Death',
    mode = 'markers',
    x = list(df1['ds']),
    y = list(df1['y']),
    marker=dict(
        color='#FFBAD2',
        line=dict(width=1)
    )
)
```

```python
trace1 = go.Scatter(
    name = 'trend',
    mode = 'lines',
    x = list(forecast['ds']),
    y = list(forecast['yhat']),
    marker=dict(
        color='red',
        line=dict(width=3)
    )
)
```

```python
upper_band = go.Scatter(
    name = 'upper band',
    mode = 'lines',
    x = list(forecast['ds']),
    y = list(forecast['yhat_upper']),
    line= dict(color='#57b88f'),
    fill = 'tonexty'
)
```

```python
lower_band = go.Scatter(
    name= 'lower band',
    mode = 'lines',
    x = list(forecast['ds']),
    y = list(forecast['yhat_lower']),
    line= dict(color='#1705ff')
)
```

```python
tracex = go.Scatter(
    name = 'Actual Death',
   mode = 'markers',
   x = list(df1['ds']),
   y = list(df1['y']),
   marker=dict(
      color='black',
      line=dict(width=2)
   )
)
```

```python
data = [tracex, trace1, lower_band, upper_band, trace]

layout = dict(title='Death prediction due to covid-19 Using FbProphet',
             xaxis=dict(title = 'Dates', ticklen=2, zeroline=True))

figure=dict(data=data,layout=layout)
```

```python
plt.savefig('forecastcovid_death.png')
```
